```python
# from the rnn page common.utils
def load_data(num):
    """Load the GEFCom 2014 energy load data"""

    sig, fields = wfdb.rdsamp(f'bidmc{num}', pb_dir='bidmc')
    b = pd.read_csv(os.path.join('data/', f'bidmc_{num}_Breaths.csv'), header=0, names=['b0','b1'])
    b.loc[:,'b2'] = (b['b0'] + b['b1'])/2

    b.drop(columns=['b0','b1'], inplace=True) 
    cols = fields['sig_name']
    
    comment = fields['comments'][0]
    
    s = re.compile(r'<age>: (\d+) <sex>: (\w)')
    res = s.search(comment)
    #age = res.group(1)
    #sex = res.group(2)
    s = pd.DataFrame(data=sig, columns=cols)
    s['breath'] = 0
    last = 0
    for v in b['b2']:
        if math.isnan(v):
            v = last
        else:
            last = v
        br = int(math.floor(v))
        s.ix[br,'breath'] = 1
    #f = pd.DataFrame({'age': [age], 'sex': [sex]})
    #s['key'] = 1
    #b['key'] = 1
    #df = pd.merge(s,b,on='key')
    #del df['key']

    # energy = pd.read_csv(os.path.join(data_dir, 'energy.csv'), parse_dates=['timestamp'])

    # Reindex the dataframe such that the dataframe has a record for every time point
    # between the minimum and maximum timestamp in the time series. This helps to 
    # identify missing time periods in the data (there are none in this dataset).
    

    return {'sig': s.ix[:,['II,','PLETH,','breath']], 'keys': cols}
```

```python
keys = set()
df = pd.DataFrame(columns=['II,','PLETH,','breath'])
for i in range(1,54):
    num = str(i).zfill(2)
    signals = None
    signals = load_data(num)
    
    keys.update(signals['keys'])
    resample = signals['sig']
    
    
    #df = df.append(signals, ignore_index=True)
    df = pd.concat([df, signals['sig']], axis=0, ignore_index=True)

    
    print(signals['sig'].tail())
#print(f'Keys: ({len(keys)}): {keys}')
```

```python
df2 = df.iloc[:, :]
df3 = pd.DataFrame()
for i in df2.index:
    if i % 25 == 0:
        A = pd.Series([df2.ix[i, 'II,'], df2.ix[i, 'PLETH,'], df2.ix[i:i+25, 'breath'].sum()])
        df3 = df3.append(A, ignore_index=True)
        #df3['breath'] = df3.append(df2.ix[i:i+25, 'II,'].sum())
df3.columns=['II','PLETH','breath']
df3.describe()
```

```python
df3.drop(df3.tail(3).index,inplace=True)
```

```python
X = df3.drop('breath', axis=1).values
y = df3['breath']
```

```python
train_len = int(len(X)*0.67)
X_train, y_train = X[:train_len], y[:train_len]
X_test, y_test = X[train_len:], y[train_len:]
```
